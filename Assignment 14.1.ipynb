```python
#Task 1

from sqlalchemy import create_engine
engine=create_engine('sqlite:///Northwind.sqlite')
from pandas import DataFrame
import pandas as pd
```

```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data")
df
```

```python
df.columns=['age','typeofjob','Empid','status','cgpa','maritalstatus','service','position','race','gender','r1','r2','r3','country','salary']
```

```python
df.head(5)
```

```python
connection=engine.connect()
```

```python
create_table_query=connection.execute("""create table data (age int,typeofjob TEXT,
Empid int,
status TEXT,
cgpa int,
maritalstatus TEXT,
service TEXT,
position TEXT,
race TEXT,
gender TEXT,
r1 int,
r2 int,
r3 int,
country TEXT,
salary int)""")
```

```python
tables=engine.table_names()
print(tables)
```

```python
df.to_sql('data', connection,if_exists='replace')
```

```python
show_table_query=connection.execute("select * from data")
table=pd.DataFrame(show_table_query.fetchall())
print(table)
```
